```python
#importing libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#to visualize with large figures
def show_image(image,cm="rainbow",shape=(6,6)):
    fig = plt.figure(figsize = shape)
    ax = fig.add_subplot(111)
    ax.imshow(image,cmap = cm)
```

```python
#reading the image and transforming it to gray scale
image = cv2.imread("brain_mri_scan.jpg")
image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
show_image(image, "gray")
```

```python
#creating the sobel filter
sobel_x = np.array([[1,0,-1]
                   ,[2,0,-2]
                   ,[1,0,-1]])

#applying the sobel edge ddetector filter
filtered = cv2.filter2D(image,-1,sobel_x)
```

```python
#creating gaussian blur filter
gbf = np.array([[1,2,1]
               ,[2,4,2]
               ,[1,2,1]])*(1/16)
filtered_gbf = cv2.filter2D(filtered,-1,gbf)
#plotting the results
fig, (ax1, ax2) = plt.subplots(1,2,figsize = (12,12))
ax1.set_title("Before")
ax1.imshow(filtered,cmap= "gray")
ax1.set_title("After")
ax2.imshow(filtered_gbf, cmap= "gray")
```

```python
#zooming in the results
fig, (ax1, ax2) = plt.subplots(1,2,figsize = (10,10))
ax1.set_title("Before")
ax1.imshow(filtered[200:400,:100],cmap= "gray")
ax1.set_title("After")
ax2.imshow(filtered_gbf[200:400,:100], cmap= "gray")
```

```python
#applying cv2 built in gaussian blur filter with 5x5 kernel 
filtered_cv_gbf = cv2.GaussianBlur(filtered,(5,5),0)
```

```python
#plotting results
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1,4,figsize = (12,12))
ax1.set_title("Original")
ax1.imshow(image,cmap= "gray")
ax2.set_title("Sobel x added")
ax2.imshow(filtered, cmap= "gray")
ax3.set_title("3x3 gaussian blur added")
ax3.imshow(filtered_gbf, cmap= "gray")
ax4.set_title("5x5 gaussian blur added")
ax4.imshow(filtered_cv_gbf, cmap= "gray")
```

```python
#zooming in the results
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1,4,figsize = (12,12))
ax1.set_title("Original")
ax1.imshow(image[200:400,:100],cmap= "gray")
ax2.set_title("Sobel x added")
ax2.imshow(filtered[200:400,:100], cmap= "gray")
ax3.set_title("3x3 gaussian blur added")
ax3.imshow(filtered_gbf[200:400,:100], cmap= "gray")
ax4.set_title("5x5 gaussian blur added")
ax4.imshow(filtered_cv_gbf[200:400,:100], cmap= "gray")
```
