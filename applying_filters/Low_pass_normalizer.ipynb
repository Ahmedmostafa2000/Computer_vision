```python
#importing libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#to visualize with large figures
def show_image(image,cm="rainbow",shape=(6,6)):
    fig = plt.figure(figsize = shape)
    ax = fig.add_subplot(111)
    ax.imshow(image,cmap = cm)
```

```python
#reading the image and transforming it to gray scale
image = cv2.imread("plane.png")
image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
show_image(image, "gray")
```

```python
#creating a low pass filter (normalizer)
low_pass_filter = np.ones((3,3))*(1/9)
filterd_1 = cv2.filter2D(image,-1,low_pass_filter)
fig, (ax1, ax2) = plt.subplots(1,2,figsize = (12,12))
ax1.set_title("Before")
ax1.imshow(image,cmap= "gray")
ax1.set_title("After")
ax2.imshow(filterd_1, cmap= "gray")
```

```python
#Zooming in to show diffrence
fig, (ax1, ax2) = plt.subplots(1,2,figsize = (12,12))
ax1.set_title("Before")
ax1.imshow(image[100:500,100:400],cmap= "gray")
ax1.set_title("After")
ax2.imshow(filterd_1[100:500,100:400], cmap= "gray")
```
