```python
#importing libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#to visualize with large figures
def show_image(image,cm="rainbow",shape=(10,10)):
    fig = plt.figure(figsize = shape)
    ax = fig.add_subplot(111)
    ax.imshow(image,cmap = cm)
```

```python
#reading the image and transforming it to gray scale
image = cv2.imread("image.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
show_image(image)
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
show_image(image,"gray")
```

```python
#creating the sobel filter
sobel_x = np.array([[1,0,-1]
                   ,[2,0,-2]
                   ,[1,0,-1]])

#applying the sobel edge ddetector filter
filterd_1 = cv2.filter2D(image,-1,sobel_x)
filterd_2 = cv2.filter2D(image,-1,np.flip(sobel_x,1))
filterd_3 = cv2.filter2D(image,-1,np.rot90(sobel_x,1,(0,1)))
filterd_4 = cv2.filter2D(image,-1,np.rot90(sobel_x,3,(0,1)))
```

```python
#plotting results
fig, (ax1, ax2) = plt.subplots(1,2,figsize = (12,12))
ax1.set_title("vertical edges detecors")
ax1.imshow(filterd_1,cmap= "gray")
ax2.imshow(filterd_2, cmap= "gray")
fig2, (ax3, ax4) = plt.subplots(1,2,figsize = (12,12))
ax1.set_title("horizontal edges detecors")
ax3.imshow(filterd_3, cmap= "gray")
ax4.imshow(filterd_4, cmap= "gray")
```
