```python
#importing libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# load in color image for face detection
image = cv2.imread('images/sherlock.jpeg')
# convert to RBG
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))
plt.imshow(image)
```

```python
# convert to grayscale
image_2 = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)  

plt.figure(figsize=(10,10))
plt.imshow(image_2, cmap='gray')
```

```python
# load in cascade classifier
face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# run the detector
faces = face_detector.detectMultiScale(image_2, 3, 1)
```

```python
len(faces)
```

```python
faces
```

```python
image_with_detections = np.copy(image)   # make a copy of the original image to plot rectangle detections ontop of

for (x,y,w,h) in faces:
    # drawing recatngles on the detected faces 
    cv2.rectangle(image_with_detections,(x,y),(x+w,y+h),(255,255,255),3)
    cv2.putText(image_with_detections,"Found one here", (x,y-10)
                , cv2.FONT_HERSHEY_SIMPLEX
                , .4
                , (255,255,255))

# showing the result
plt.figure(figsize=(10,10))
plt.imshow(image_with_detections)
```
