```python
#importing libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#to visualize with large figures
def show_image(image,cm="rainbow"):
    fig = plt.figure(figsize = (10,10))
    ax = fig.add_subplot(111)
    ax.imshow(image,cmap = cm)
```

```python
#importing and showing image
image = cv2.imread("batman.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
show_image(image)
```

```python
#creating a colour threshold 
mask = cv2.inRange(image,(160,0,195),(180,255,255))
show_image(mask,"gray")
#applying mask
masked_im = np.copy(image)
masked_im[mask != 0] = [0,0,0]
masked_im = cv2.cvtColor(masked_im,cv2.COLOR_HSV2RGB)
show_image(masked_im)
```

```python
background = cv2.imread("Gotham.jpg")
background = cv2.cvtColor(background,cv2.COLOR_BGR2RGB)
background = cv2.resize(background,(728,455))
show_image(background)
```

```python
background[mask== 0] = [0,0,0]
show_image(background)
```

```python
complete = masked_im + background
show_image(complete)
```
