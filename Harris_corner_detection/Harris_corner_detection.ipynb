```python
#importing libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#reading the image
image = cv2.imread("image.jpg")
image = image[:600,:]
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
#transforming the image to gray scale
image_2 = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(image,cmap = "gray")
```

```python
#using canny detection to detect edges
edges = cv2.Canny(image_2,100,255)
corners = cv2.cornerHarris(edges, 2, 3, .1)
plt.imshow(corners,cmap = "gray")
```

```python
#dilating corners
corners = cv2.dilate(corners, None) 
plt.imshow(corners, cmap = "gray")
```

```python
#showing results
corners_copy = np.copy(image)
threshold = 0.1*corners.max()
for i in range(corners.shape[0]):
    for j in range(corners.shape[1]):
        if corners[i,j] > threshold:
            cv2.circle(corners_copy,(j,i), 2, (0,255,255), 2)

fig, ax1 = plt.subplots(figsize = (10,10))
ax1.imshow(corners_copy)
```
