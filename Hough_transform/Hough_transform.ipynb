```python
#importing libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#reading the image and transforming it to gray scale
image = cv2.imread("laptop.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
#applying Canny edge detection
image2 = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
edges = cv2.Canny(image2,130,255)
plt.imshow(edges,cmap = "gray")
```

```python
#creating the hough 
lines = cv2.HoughLinesP(edges, 1, np.pi/180, 20, 1000, 15)
image_with_lines = np.copy(image)
for line in lines:
    for x1, y1, x2, y2 in line:
        cv2.line(image_with_lines,(x1,y1),(x2,y2),(0,255,0),4)
fig, ax1 = plt.subplots(1,1,figsize = (8,8))
ax1.imshow(image_with_lines,cmap= "gray")
```
