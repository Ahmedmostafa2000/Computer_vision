```python
#importing libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#to visualize with large figures
def show_image(image,cm="rainbow",shape=(6,6)):
    fig = plt.figure(figsize = shape)
    ax = fig.add_subplot(111)
    ax.imshow(image,cmap = cm)
```

```python
#reading the image and transforming it to gray scale
image = cv2.imread("tesla.jpg")
image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
show_image(image, "gray")
```

```python
#Applying the threshold
ret,thresh1 = cv2.threshold(image,80,255,cv2.THRESH_BINARY)
show_image(thresh1,"gray")
```

```python
ret,thresh2 = cv2.threshold(image,127,255,cv2.THRESH_TRUNC)
show_image(thresh2,"gray")
```
