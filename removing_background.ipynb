```python
#importing libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#to visualize with large figures
def show_image(image,cm="rainbow"):
    fig = plt.figure(figsize = (10,10))
    ax = fig.add_subplot(111)
    ax.imshow(image,cmap = cm)
```

```python
#importing and showing image
image = cv2.imread("maxresdefault.jpg")
show_image(image)
```

```python
#creating a colour threshold 
mask = cv2.inRange(image,(0,245,0),(255,255,255))
show_image(mask,"gray")
```

```python
#applying mask
masked_im = np.copy(image)
masked_im[mask != 0]=[0,0,0]
show_image(masked_im)
```
